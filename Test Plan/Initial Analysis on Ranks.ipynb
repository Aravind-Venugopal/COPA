```python
res = res.drop(res.loc[res['Panel'].str.contains('Please write any additional comments')].index, axis=0)
res = res.drop(res.loc[res['Panel'].str.contains('When looking')].index, axis=0)
```

```python
res = res.reset_index(drop=True)
res['PanelType'] = reference_df['Panel Type']
res['SearchString'] = reference_df['0']
res['PanelType'] = res['PanelType'].apply(lambda x: x.split(' ')[0])
```

```python
pivot_res = pd.melt(res, id_vars=['Panel', 'PanelType', 'SearchString'])
pivot_res['value'] = pivot_res['value'].astype(int)
```

```python
pivot_res['Rank'] = pivot_res.groupby(['SearchString', 'variable'])['value'].rank(ascending=False)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
pivot_res.head()
```

```python
model = ols('value ~ C(PanelType) + C(Rank) + C(PanelType):C(Rank)', data=pivot_res).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
model = ols('value ~ C(PanelType) + C(variable) + C(PanelType):C(variable)', data=pivot_res).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

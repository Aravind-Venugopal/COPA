```python
import pandas as pd
import re
import numpy as np
```

```python
Matrix = pd.read_csv('FrankMatrix.csv')
Matrix = Matrix[['Article','Author','Content','Date','Source','topthr']]
```

```python
#Source (30%) + Time (20%) + Top Three Words (50%) = 100 scores
#Three Different sources: WIKI is 10 points, Blog is 20 points and Mag is 30 points
#Time: 2019 is 20 points, 2018 is 17 points, 2017 is 14 points .....
#Top Three Words: If search words hit one of this three words, get 50 points
#Author: Is valiable, but due to the lack of authors in here, will not included

def POCfind(intext):
 n = len(Matrix)
 df = [pd.DataFrame() for x in range(n)]
 for i in range(n):
   df[i]=re.search(intext,Matrix.Content[i])

 ds = [pd.DataFrame() for x in range(n)]
 for i in range(n):
   ds[i]=re.search(intext,str(Matrix.topthr[i]))
 ds = pd.DataFrame(ds)
 ds = np.where(ds)
 return np.where(df), ds
   
```

```python
### Put the word you want to search in here
test = POCfind('copa')
```

```python
#Cal the source score

n= len(test[0][0])-1
SourceS = [pd.DataFrame() for x in range(n)]

for i in range(n):
 SourceS[i] = Matrix.get_value(int(test[0][0][i]),'Source')*10
```

```python
SourceS = pd.DataFrame(SourceS)
Test = pd.DataFrame(test[0][0])
Test = Test.rename(columns={Test.columns[0]:'Index'})
Score = pd.concat([Test,SourceS],axis=1)
Score = Score.rename(columns={Score.columns[1]:'SScore'})
```

```python
#Cal the date score
n= len(test[0][0])-1
DateS = [pd.DataFrame() for x in range(n)]

for i in range(n):
 DateS[i] = 20-(2019-int(Matrix.get_value(int(test[0][0][i]),'Date')))*3
```

```python
DateS = pd.DataFrame(DateS)
Score = pd.concat([Score,DateS],axis=1)
Score = Score.rename(columns={Score.columns[2]:'DScore'})
```

```python
#Cal the top thr world 
# n=len(Matrix)
# ds = [pd.DataFrame() for x in range(n)]
# for i in range(n):
#     ds[i]=re.search('pilot',str(Matrix.topthr[i]))
#    # ds[i] = ds[i].group()
# ds = pd.DataFrame(ds)
# ds = np.where(ds)

n= len(test[1][0])
topS = [pd.DataFrame() for x in range(n)]
for i in range(n):
  topS[i] = 50

topS = pd.DataFrame(topS)
Test = pd.DataFrame(test[1][0])
Test = Test.rename(columns={Test.columns[0]:'Index'})
tcore = pd.concat([Test,topS],axis=1)
tcore = tcore.rename(columns={tcore.columns[1]:'tcore'})
```

```python
final_score = pd.merge(left=Score,right=tcore, how='left', left_on='Index', right_on='Index')
final_score = final_score.fillna(0)
final = final_score['SScore'].astype(int)+final_score['DScore'].astype(int)+final_score['tcore'].astype(int)
final = pd.DataFrame(final)
final = final.rename(columns={final.columns[0]:'Final'})
```
